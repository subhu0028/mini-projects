```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
data = pd.read_csv('/kaggle/input/heart-disease/heart.csv')
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
import seaborn as sns
correlation_mat = data.corr()
top_corr_features = correlation_mat.index
plt.figure(figsize=(16,16))
#plot heat map
g=sns.heatmap(data[top_corr_features].corr(),annot=True,cmap="RdYlGn")
plt.show()
```

```python
sns.set_style('whitegrid')
sns.countplot(x='target',data=data,palette='RdBu_r')
plt.show()
```

```python
dataset = pd.get_dummies(data, columns = ['sex', 'cp', 'fbs',
                                          'restecg', 'exang',  
                                         'slope', 'ca', 'thal'])
dataset.head()
```

```python
dataset.describe()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
standardScaler = StandardScaler()
columns_to_scale = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
dataset[columns_to_scale] = standardScaler.fit_transform(dataset[columns_to_scale])
dataset.head()
```

```python
y = dataset['target']
x = dataset.drop(['target'], axis=1)
```

```python
from sklearn.model_selection import cross_val_score
knn_scores = []
for k in range(1, 21):
    knn_classifier = KNeighborsClassifier(n_neighbors = k)
    score = cross_val_score(knn_classifier, x, y, cv=10)
    knn_scores.append(score.mean())
```

```python
knn_scores
```

```python
plt.plot([k for k in range(1, 21)], knn_scores, color = 'blue')
plt.xticks([i for i in range(1, 21)])
plt.xlabel('Number of Neighbors (K)')
plt.ylabel('Scores')
plt.title('K Neighbors Classifier scores for different K values')
plt.show()
```

```python
knn_scores[11]
```

```python
from sklearn.ensemble import RandomForestClassifier
randomForestClassifier = RandomForestClassifier(n_estimators = 10)
score = cross_val_score(randomForestClassifier, x, y, cv=10)
```
